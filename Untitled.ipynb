```python
#
lst = []
for row in today_df.itertuples():
    lst.append(list(row[1:]))
    lst += [list(r[1:]) for r in pre_df.itertuples() if row.No == r._1]
col = ["No", "machine", "date", "", "", "", ""]
merge_df = pd.DataFrame(lst, columns=col).sort_values(['No', 'date']).reset_index(drop=True)
merge_df.head(10)
```

```python
# 
dic = {}
for idx, row in merge_df.iterrows():
    dic.update({idx: list(row)})
    
dt_now = datetime.datetime.now()
oneday = datetime.timedelta(days=1)
if dt_now.hour < 12: dt_now -= oneday
yyyymmdd = dt_now.strftime('%Y-%m-%d')

read_json = json.dumps(dic, ensure_ascii=False)
p_w = f"./ds_zerogame_{yyyymmdd}.json"
with open(p_w, "w", encoding="utf-8") as f:
    f.write(read_json)
```

```python
import tkinter as tk

class TkDf(tk.Tk):

    def __init__(self, df: pd.DataFrame):
        super().__init__()
        self.title("Display DataFrame")
        self.col_line = 0

        if type(df.columns) == pd.core.indexes.base.Index:
            for c in range(len(df.columns)):
                e = tk.Entry(self)
                e.insert(0, df.columns[c])
                e.grid(row=0, column=c)
            self.col_line = 1
            
        for r in range(len(df)):
            for c in range(len(df.columns)):
                e = tk.Entry(self)
                e.insert(0, df.iloc[r,c])
                e.grid(row=r+self.col_line,column=c)
```

```python
df = pd.DataFrame([[1,2,3], [4,5,6]])
df.columns = ["a", "b", "c"]
TkDf(df).mainloop()
```
